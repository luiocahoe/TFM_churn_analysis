# Machine Learning para predecir cancelaciones y mejorar la retención en seguros

## Modelos de machine learning

```python
import pandas as pd
X_test = pd.read_csv('X_test.csv')
y_test = pd.read_csv('y_test.csv').squeeze()
X_train_ros = pd.read_csv('X_train_ros.csv')
y_train_ros = pd.read_csv('y_train_ros.csv').squeeze()
X_train_rus = pd.read_csv('X_train_rus.csv')
y_train_rus = pd.read_csv('y_train_rus.csv').squeeze()
X_train_smote = pd.read_csv('X_train_smote.csv')
y_train_smote = pd.read_csv('y_train_smote.csv').squeeze()
```

```python
from sklearn.metrics import classification_report, roc_auc_score, f1_score
```

```python
from sklearn.linear_model import LogisticRegression

# Modelo con RandomOverSampler
logistic_ros = LogisticRegression(max_iter=1000, random_state=42)
logistic_ros.fit(X_train_ros, y_train_ros)
y_pred_ros = logistic_ros.predict(X_test)
y_proba_ros = logistic_ros.predict_proba(X_test)[:, 1]
print("\nResultados para RandomOverSampler:")
print(f"AUC: {roc_auc_score(y_test, y_proba_ros):.4f}")
print(f"F1-score: {f1_score(y_test, y_pred_ros):.4f}")
print(classification_report(y_test, y_pred_ros))

# Modelo con RandomUnderSampler
logistic_rus = LogisticRegression(max_iter=1000, random_state=42)
logistic_rus.fit(X_train_rus, y_train_rus)
y_pred_rus = logistic_rus.predict(X_test)
y_proba_rus = logistic_rus.predict_proba(X_test)[:, 1]
print("\nResultados para RandomUnderSampler:")
print(f"AUC: {roc_auc_score(y_test, y_proba_rus):.4f}")
print(f"F1-score: {f1_score(y_test, y_pred_rus):.4f}")
print(classification_report(y_test, y_pred_rus))

# Modelo con SMOTE
logistic_smote = LogisticRegression(max_iter=1000, random_state=42)
logistic_smote.fit(X_train_smote, y_train_smote)
y_pred_smote = logistic_smote.predict(X_test)
y_proba_smote = logistic_smote.predict_proba(X_test)[:, 1]
print("\nResultados para SMOTE:")
print(f"AUC: {roc_auc_score(y_test, y_proba_smote):.4f}")
print(f"F1-score: {f1_score(y_test, y_pred_smote):.4f}")
print(classification_report(y_test, y_pred_smote))
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Input
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential

def crear_modelo_red_neuronal(X_train, y_train, X_val, y_val, epochs=100, batch_size=256):
    # Crear el modelo
    model = Sequential([
        Input(shape=(X_train.shape[1],)),
        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.3),
        Dense(32, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),
        Dense(1, activation='sigmoid')
    ])

    # Compilar el modelo
    optimizer = Adam(clipvalue=1.0)  # Usando gradient clipping
    model.compile(optimizer=optimizer,
                  loss='binary_crossentropy',
                  metrics=['accuracy', tf.keras.metrics.AUC(name='auc')])

    # Definir los callbacks
    early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-6, verbose=1)

    # Entrenamiento del modelo
    history = model.fit(X_train, y_train,
                        validation_data=(X_val, y_val),
                        epochs=epochs,
                        batch_size=batch_size,
                        callbacks=[early_stop, reduce_lr],
                        verbose=1)

    return history, model
```

```python
from sklearn.metrics import classification_report, roc_auc_score, f1_score
from sklearn.model_selection import train_test_split

# Red Neuronal con RandomOverSampler
X_ros_train, X_ros_val, y_ros_train, y_ros_val = train_test_split(
    X_train_ros, y_train_ros, test_size=0.2, stratify=y_train_ros, random_state=42
)
history_ros, model_ros = crear_modelo_red_neuronal(X_ros_train, y_ros_train, X_ros_val, y_ros_val)

# Red Neuronal con RandomUnderSampler
X_rus_train, X_rus_val, y_rus_train, y_rus_val = train_test_split(
    X_train_rus, y_train_rus, test_size=0.2, stratify=y_train_rus, random_state=42
)
history_rus, model_rus = crear_modelo_red_neuronal(X_rus_train, y_rus_train, X_rus_val, y_rus_val)

# Red Neuronal con SMOTE
X_smote_train, X_smote_val, y_smote_train, y_smote_val = train_test_split(
    X_train_smote, y_train_smote, test_size=0.2, stratify=y_train_smote, random_state=42
)
history_smote, model_smote = crear_modelo_red_neuronal(X_smote_train, y_smote_train, X_smote_val, y_smote_val)
```
